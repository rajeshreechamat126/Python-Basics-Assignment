Que3. Compare and contrast map, filter, and reduce.

Ans:
     The differences between map, filter and reduce are:
     
     1.map(): The map() function is a type of higher-order. This function takes another function as parameter along with a          sequence of iterables and returns an output after applying the function to each iterable present in the sequence. 
     
     2.filter(): The filter() function is used to create an output list consisting of values for which the function returns          true.
     
     3.reduce(): The reduce () function, as the name describles, applies a given function to the iterables and returns a single      value.
     

```python
### Map functions 
Animal = ["Anteater", "Dog", "Elephant", "Giraffe", "Cat"]
map_object = map(lambda s: s[0] == "A", Animal)

for i in map_object:
    print(i)
    
```

```python
### Filter function
Animal = ["Anteater", "Dog", "Elephant", "Giraffe", "Cat"]
filter_object = filter(lambda s: s[0] == "A", Animal)

for i in filter_object:
    print(i)
```

```python
### Reduce Function 
from functools import reduce 

list = [4, 3, 8, 3] 
print(reduce(lambda x, y: x + y, list))
print("with an initial value: " + str(reduce(lambda x, y: x + y, list, 10)))
```

Que4. What are function annotations, and how are they used?

Ans: 
      Function annotation is the standard way to access the metadata with the arguments and the return value of the function.         These are nothing but some random and optional Python expressions that get allied to different parts of the function.           They get evaluated only during the compile-time and have to significance during the run-time of the code. They do not           have any signficance or meaning associated with them until accessed by some third- party libraries. They are used to type       check the functions by declaring the type of the parameters and the return value for the functions. The string-based           annotations help us to improve the help messages. 
      
      Syntax:
      
      def func(a:'int')->'int':
      
      pass
      
      annotations for simple parameters:
      
      def func(x:'float'=10.8,y:'argument2'):
      
      In the above code the argument,'x' of the function func, has been annotated to float data type and the argument'y' has a       string-based annotation. The argument can also be assigned to a default value using a'='symbol followed by the default         value. These default values are optional to the code. 
      
      Annotations for return values:
      
      def func(a:expression)->'int':
      
      The annotations for the return value is written after the '->'symbol.
      

Que5. What are recursive functions, and how are they used?

Ans:
     A recursive function is a function that calls itself during its execution. This means that the function will continue to        call itself and repeat its behavior until some condition is met to return a result.
     

```python
# Example
def fact(x):
    if x == 1 :
        return 1
    else :
         return x * fact(x-1) # recurtion
    
fact(3)
```
